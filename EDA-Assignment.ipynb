### FINDING THE TOP 10 CO-RELATIONS IN TARGET ZERO AND TARGET ONE DATAFRAMES

```python
#Finding the top 10 Co-relations in TARGET_ZERO dataset.

corr_matrix = TARGET_ZERO.corr().abs()

#the matrix is symmetric so we need to extract upper triangle matrix without diagonal (k = 1)

corr_TARGET_ZERO=corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(np.bool))
corr_TAREGT_ZERO_df=corr_TARGET_ZERO.unstack().reset_index()
corr_TAREGT_ZERO_df.columns=["VAR1","VAR2","CORELATION"]
corr_TAREGT_ZERO_df.dropna(subset=["CORELATION"],inplace=True)
corr_TAREGT_ZERO_df["CORELATION"]=corr_TAREGT_ZERO_df["CORELATION"].apply(lambda x: round(x,2))
corr_TAREGT_ZERO_df.sort_values(by="CORELATION", ascending=False).head(10)

```

```python
#Finding the top 10 Co-relations in TARGET_ONE dataset

corr_matrix1 = TARGET_ONE.corr().abs()

#the matrix is symmetric so we need to extract upper triangle matrix without diagonal (k = 1)

corr_TARGET_ONE=corr_matrix1.where(np.triu(np.ones(corr_matrix1.shape),k=1).astype(np.bool))
corr_TARGET_ONE_df=corr_TARGET_ONE.unstack().reset_index()
corr_TARGET_ONE_df.columns=["VAR1","VAR2","CORELATION"]
corr_TARGET_ONE_df.dropna(subset=["CORELATION"],inplace=True)
corr_TARGET_ONE_df["CORELATION"]=corr_TARGET_ONE_df["CORELATION"].apply(lambda x: round(x,2))
corr_TARGET_ONE_df.sort_values(by="CORELATION", ascending=False).head(10)
```
